```python
import re
def CalTF(aname, alist):
  sum = 0
  aname = re.sub(r'[^\w]', '', aname)
  aname = re.sub(r'[0-9]', '', aname)
  for i in range(len(alist)):
    string = alist[i]
    substring = aname
    sum += string.count(substring)
  return sum
```

```python
def CalDF(aname, alist):
  cnt = 0
  aname = re.sub(r'[^\w]', '', aname)
  aname = re.sub(r'[0-9]', '', aname)
  for i in range(len(alist)):
    string = alist[i]
    substring = aname
    if(string.find(substring) != -1):
      cnt += 1
  return cnt
```

```python
stock_TF = dict()
for stock_name in name_list:
  TF = CalTF(stock_name, text_list)
  if (TF > 10000):
    stock_TF[stock_name] = TF
```

```python
stock_DF = dict()
for stock_name in name_list:
  DF = CalDF(stock_name, text_list)
  if (DF > 5000):
    stock_DF[stock_name] = DF
```

```python
# 針對每個個股，計算其漲跌幅度
# 計算方式：
# 1) 將價格差異紀錄下來，並且計算平均
# 2) 將價格差異與平均的差取平方相加，除以價格差異清單長度

def CalVar(this_name):

  price_list = []
  for index, row in df_2018.iterrows():
    aname = row['證券代碼']
    aprice = row['收盤價(元)']
    if aname == this_name:
      price_list.append(float(aprice))

  sr = pd.Series(price_list)
  var_price = sr.var()/len(price_list)

  return var_price
```

## 漲跌文章標記

```python
import datetime
def rf_lists(n, set_perc, all_list):
  rise_articles = []
  fall_articles = []

  for i in range(len(all_list)):
    today = all_list[i][0]
    content = all_list[i][1]

    # today
    while (today not in stock_data_dict.keys() and today < max_day):
      today += datetime.timedelta(days = 1)

    # after
    cnt = 0
    add_day = 0
    today_temp = today
    while (cnt < n):
      if (today_temp < max_day):
        today_temp += datetime.timedelta(days = 1)
        if (today_temp in stock_data_dict.keys()):
          cnt += 1
          add_day += 1
        else: #weekend
          add_day += 1
      else:
        break
    time_delta = datetime.timedelta(days = add_day)
    af_day = today + time_delta

    # print("today: ", today)
    # print("day+3: ", af_day)
    if (today < max_day - datetime.timedelta(days = n+1)):
      today_price = stock_data_dict[today]
      future_price = stock_data_dict[af_day]
      percentage = ((future_price - today_price) / today_price)*100
      # print("percentage", percentage)

      # clean text
      string = content
      clean_str = re.sub(r'[^\w]', '', string)
      text_str = re.sub(r'[A-Za-z0-9]', '', clean_str)

      if (percentage >= set_perc):
        rise_articles.append(text_str)
      elif (percentage <= -set_perc):
        fall_articles.append(text_str)

  return rise_articles, fall_articles
```

## Functions (切詞選詞)

```python
# 將每篇文章依據n gram切詞，並計算各個gram的TF

def ngramoperator(astring, n):
    chlist=[]
    for i in range(len(astring) - n + 1):  # 由n決定迴圈圈數
        chlist.append(astring[i : i + n])  # 決定一次切n個字
    
    freqdict = dict()
    for ch in chlist:
        freqdict[ch] = freqdict.get(ch, 0) + 1
    
    return freqdict


from operator import itemgetter

# 將得到的gram dict依據TF做排序，回傳一個list

def sort(adict):
    freqsorted = sorted(adict.items(), key=itemgetter(1), reverse=True)

    return freqsorted
```

```python
# 計算各個gram的DF

def DF_count(gram, yourdata):
	df_count = 0

	for i in yourdata:
		if i.find(gram) != -1:
			df_count += 1

	return df_count
```

```python
def data2ngram(adata, n):

  adict = dict()
  
  # 將每篇文章切成一個gram list並計算gram freq
  for i in range(len(adata)):
    test = str(adata[i])
    chfreqdict = ngramoperator(test, n)
    chfreqsorted = sort(chfreqdict)

    # 比對新的gram list與總清單，更新已有的gram freq或是加入新的gram
    for j in range(len(chfreqsorted)):
      word = chfreqsorted[j][0]
      if word in adict:
        adict[word] += chfreqsorted[j][1]
      else:
        adict[word] = chfreqsorted[j][1]

  dictsorted = sort(adict)  # 將得到的gram dict依據TF值做排序

  ngram_TF = dictsorted[:1000]  # 決定要抓多少gram(全部的抓10000個，分主題抓100就好)
  ngram_num = len(dictsorted)

  # 依據計算TF後得到的一組gram，計算其DF值
  for i in range(len(ngram_TF)):
    gram = ngram_TF[i][0]
    counts = DF_count(gram, adata)
    L1 = list(ngram_TF[i])
    L1.append(counts)
    ngram_TF[i] = L1

  return ngram_TF
```

```python
# Substring Process

def Substring_ttt(agramlist, bgramlist):
  to_del_list = []
  for agram, atf, adf in agramlist:
    x = 0
    for bgram, btf, bdf in bgramlist:
      if agram in bgram and ((abs(adf - bdf)/adf) <= 0.01):
        to_del_list.append([agram, atf, adf])
        #agramlist.remove(agram, atf, adf))
        x += 1 
        if x == 2:
          break
  #print(to_del_list)
  for delgram, deltf, deldf in to_del_list:
    for agram, atf, adf in agramlist:
      if agram in delgram:
        agramlist.remove([agram, atf, adf])
  
  return agramlist 

```

```python
def get_ngram_words(articles):
  bigram = data2ngram(articles, 2)
  trigram = data2ngram(articles, 3)
  four_gram = data2ngram(articles, 4)

  bigram = Substring_ttt(bigram, trigram)
  trigram = Substring_ttt(trigram, four_gram)

  gram_dict = dict()
  for i in range(len(bigram)):
    bi_name = bigram[i][0]
    gram_dict[bi_name] = bigram[i][1] / bigram[i][2]
  for i in range(len(trigram)):
    tri_name = trigram[i][0]
    gram_dict[tri_name] = trigram[i][1] / trigram[i][2]
  for i in range(len(four_gram)):
    four_name = four_gram[i][0]
    gram_dict[four_name] = four_gram[i][1] / four_gram[i][2]

  #gram dict
  words = sorted(gram_dict.items(), key=itemgetter(1), reverse=True)
  return words
```

## 刪除重複關鍵詞

```python
def remove_same(rise_words, fall_words):
  word_to_del = set()
  for i in range(len(rise_words)):    
    for j in range(len(fall_words)):
      if (rise_words[i][0] == fall_words[j][0]):
        word_to_del.add(rise_words[i][0])

  rise_clean = []
  fall_clean = []
  for i in range(len(rise_words)):
    if (rise_words[i][0] in word_to_del):
      continue
    rise_clean.append([rise_words[i][0], rise_words[i][1]])

  for i in range(len(fall_words)):
    if (fall_words[i][0] in word_to_del):
      continue
    fall_clean.append([fall_words[i][0], fall_words[i][1]])

  return rise_clean, fall_clean
```

```python
import datetime
import calendar

def add_months(sourcedate, months):
    month = sourcedate.month - 1 + months
    year = sourcedate.year + month // 12
    month = month % 12 + 1
    day = min(sourcedate.day, calendar.monthrange(year,month)[1])
    return datetime.date(year, month, day)
```

```python
import datetime
def get_rf(n, set_perc, adate):

  today = adate
  # today
  while (today not in stock_data_dict.keys() and today < max_day):
    today += datetime.timedelta(days = 1)

  # after
  cnt = 0
  add_day = 0
  today_temp = today
  while (cnt < n):
    if (today_temp < max_day):
      today_temp += datetime.timedelta(days = 1)
      if (today_temp in stock_data_dict.keys()):
        cnt += 1
        add_day += 1
      else: #weekend
        add_day += 1
    else:
      break
  time_delta = datetime.timedelta(days = add_day)
  af_day = today + time_delta

  if (today < max_day - datetime.timedelta(days = n+1)):
    today_price = stock_data_dict[today]
    future_price = stock_data_dict[af_day]
    percentage = ((future_price - today_price) / today_price)*100
    # print("percentage", percentage)


    if (percentage > 0):
      return 1 #rise
    elif (percentage < 0):
      return -1 #fall
    else:
      return 0 #none
```
